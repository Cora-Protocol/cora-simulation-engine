```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

import seaborn as sns

sns.set_style("whitegrid")

BASE_PATH = Path("studies/003-backtesting")
```

```python
def load_csvs(path: Path) -> pd.DataFrame:
    return pd.concat([pd.read_csv(file, index_col=0) for file in path.glob("*.csv")], axis=0, ignore_index=True)
```

```python
df = load_csvs(BASE_PATH / "cache")
```

```python
subset = df[df["max_ltv"] == 0.5]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
sns.violinplot(x="utilization_parameter", y="average_utilization", hue="loan_duration_type", data=subset, ax=ax)
```

```python
subset = df[(df["max_ltv"] == 0.5 )& (df["average_utilization"] > 0.28)& (df["average_utilization"] < 0.32)]

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
sns.scatterplot( x="average_utilization", y="utilization_parameter", hue="loan_duration_type", data=subset, ax=ax)
```
